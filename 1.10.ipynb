```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from mpl_toolkits.mplot3d import Axes3D
import pylab
%matplotlib inline
```

```python
n = 100
Tstart = 1000
Tend = 10000
```

```python
c = 3e10
aem = 1.66054e-27
evToT = 11604.52500617
h = 6.626e-34
pi = 3.14
P = 101325 * 0.1
k = 1.380648e-23
ma = 35.5 * aem
mb = 35.5 * aem
h_ = h / (2 * pi)
D = 2.4975 * evToT
B = 0.2441 * h * c
mu = 17.73 * aem
We = 559.7 * h * c
ga = 1e31
gb = 6
gab = 1
```

```python
def StoZ(S):
    Z = m.exp(S / k)
    #print('stat')
    #print(u)
    return(Z)
def Stat1():
    u = ga*gb/gab
    return(u)
def Stat(Za, Zb, Zab):
    u = (Za * Zb) / Zab
    #print('stat')
    #print(u)
    return(u)
def TempMass(T):
    u = ((ma * mb) / (ma + mb) * T * k / (2 * pi * (h_ ** 2)))**(3 / 2)
    #print('TempMass')
    #print(u)
    return(u)
def Vib(T):
    u = (B / P) * (h * We / (T * k))
    #print('Vib')
    #print(u)
    return(u)
def Diss(T):
    u = m.exp(- D / (T))
    #print('Diss')
    #print(u)
    return(u)
```

```python
X = np.linspace(Tstart , Tend , n)
y = []
vert = np.linspace(0, 1, 10)
z = 0
gor = []
for T in X:
    f = m.sqrt(Stat1() * TempMass(T) * Vib(T) * Diss(T))
    #print(f)
    y.append(f / (1 + f))
    if(f/(1+f)-0.5 < 1e-5):
        z = T
for x in vert:
    gor.append(z)
```

```python
    
fig = plt.figure()
ax1 = fig.add_axes([0.125, 0.547727, 2, 2]) 
ax1.set_title(u'Аналитическое решение модельной задачи')

ax1.yaxis.set_label_position('left')
ax1.set_ylabel(u'Ось y')

ax1.xaxis.set_label_position('bottom')
ax1.set_xlabel(u'Ось x')

#for ax1 in fig.axes:
ax1.grid(True)

ax1.plot(X, y, 'g', gor, vert, 'r')
```
